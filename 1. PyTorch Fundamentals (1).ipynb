##### What are neurons? Neural Network
Consider neuron as a thing that holds number from 0 to 1. If an image is a 28x28 pixel then 28x28 = 784 neurons are created. each neurons holds number that denotes gray scale value of the corresponding pixel (ranging from 0 to black and 1 for white pixel) . the number inside the neuron is called activation value. All these 784 neurons form the first layer of the neural network. The last layer denotes the options / choices of what the output class the input image can be. For example if the neural network is build to detect if an hand written image is 9. then the output has 0 to 9 totally 10 neurons.

Activations in one layer determines, activations in second layer. if we feed in an image, lightening up all 784 neurons. According to the brightness of each pixel, an pattern is formed in the second layer. which when lighten up forms a specific pattern in second layer. this second layer on lightening up forms a specific pattern. The brightened neuron in the last layer denotes the output the model has detected it to be (here 9 is lightened up) denoting the handwritten image is 9. 

##### What are layers in neural network:



Consider in detection of image, the first layer detects the small small portion that forms the loop and straighline. The second layer then detectes if loop and line/ then final layer combines both and detects that it is 9

### What is a gradient descent !?

###### COST FUNCTION: 

Initially setting up the bias and the weights randomly. weights denotes the strength of the connections(lines) between two layer of neurons bias denotes if the activation is ON or OFF. initially we set up both bias and weights randomly. So when we ran the model, the prediction will be bad. 

The cost function of a neural network will be the sum of errors in each layer.


When the model correctly classifies , then the sum of squared difference of predicted and required output denoting cost function will be low. 





When the model doesnt classify correctly. then the cost function will be high



##### GRADIENT DESCENT:
In order for the network to learn and produce appropriate output. We adjust the weights and bias of the network such that the cost function is reduced and we get predictions values that are close to the output and not the random values that we obtained in the starting step. From changing of weights to reduce the cost function, we also know that which weight is important to reduce the cost function

##### BACK PROPOGATION:
Training a model is just minimising the loss function, and to minimise you want to move in the negative direction of the derivative. Back-propagation is the process of calculating the derivatives and gradient descent is the process of descending through the gradient, i.e. adjusting the parameters of the model to go down through the loss function.

#### Pytorch Reproducibility

When using rand operation in torch, random numbers are created everytime the notebook is run. To limit the randomness, random seed comes in picture. reproducible means whenever the notebook is run, we get same value for torch.rand

```python
torch.rand(3,3)
```

```python
import torch

a = torch.rand(3,4)
b = torch.rand(3,4)
print(a,b)
print(a == b)
```

```python
random_seed = 43

torch.manual_seed(random_seed)
c = torch.rand(3,4)

random_seed = 44

torch.manual_seed(random_seed)
d = torch.rand(3,4)

print(c,d)
print(c == d)


```
